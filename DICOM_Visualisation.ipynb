## Importing Libraries 

```python
import matplotlib.pyplot as plt
import pydicom as dicom 
import os
```

## Reading a dicom file 

```python
slice = dicom.read_file("FLAIR/IM-0001-0011.dcm")
```

## Printing the dicom slice

```python
#It contains many fields, protected health information, hospital-related meta-data
print(slice)
```

## Visualising the dicom slice 

```python
plt.imshow(slice.pixel_array, plt.cm.bone)
plt.show
```

## Checking PixelSpacing

## Checking PixelAngle 

```python
(slice.FlipAngle,)
```
